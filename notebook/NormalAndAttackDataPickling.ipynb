```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plts
```

```python
DATAFILEPATH = "../../../Data/SwaT.A1&A2-Dec2015/Physical/"
ATTACKFILEPATH = "../../../Data/SwaT.A1&A2-Dec2015/drive-download-20220826T152933Z-001/"
NORMALDATAFILE = "SWaT_Dataset_Normal_v1.xlsx"
ATTACKDATAFILE = "SWaT_Dataset_Attack_v0.xlsx"
ATTACKDETAILS = "List_of_attacks_Final.xlsx"
NORMALDATAPickleFile = "SWAT_DATA_NORMAL.pkl"
ATTACKDATAPickleFile = "SWAT_DATA_ATTACK.pkl"
ATTACKDATACleanedPickleFile = "SWAT_DATA_ATTACK_CLEANED.pkl"
NORMALDATACleanedPickleFile = "SWAT_DATA_NORMAL_Cleaned.pkl"
```

```python
ACTUALATTACKDATA = os.path.join(DATAFILEPATH , ATTACKDATAFILE)
ATTACKDETAILS = os.path.join(ATTACKFILEPATH , ATTACKDETAILS)
NORMALDATA = os.path.join(DATAFILEPATH , NORMALDATAFILE)
NORMALDATAPickle = os.path.join(DATAFILEPATH , NORMALDATAPickleFile)
ATTACKDATAPickle = os.path.join(DATAFILEPATH , ATTACKDATAPickleFile)
ATTACKDATACleaned = os.path.join(DATAFILEPATH , ATTACKDATACleanedPickleFile)
NORMALDATACleaned = os.path.join(DATAFILEPATH , NORMALDATACleanedPickleFile)

print(ACTUALATTACKDATA)
print(ATTACKDETAILS)
print(NORMALDATA)
print(NORMALDATAPickle)
print(ATTACKDATAPickle)
print(ATTACKDATACleaned)
print(NORMALDATACleaned)
```

## Create Cleaned Pickle file for Attack Data

```python
# Read from raw Pickle
dfAttackData =  pd.read_pickle(ATTACKDATAPickle)

print(dfAttackData.shape)
dfAttackData.head()
```

```python
# Get All columns - That is first row and relace column headers
colList = dfAttackData.iloc[0].tolist()
print(colList)
```

```python
dfAttackData.columns = colList
print(dfAttackData.columns)
```

```python
# rename Timestap column name
dfAttackData = dfAttackData.rename(columns={' Timestamp': 'Timestamp'})
print(dfAttackData.columns)
dfAttackData.head()
```

```python
dfAttackData = dfAttackData.iloc[1:,:]
print(dfAttackData.shape)
dfAttackData.head()
```

```python
dfAttackData.reset_index(drop=True,inplace=True)
print(dfAttackData.shape)
dfAttackData.head()
```

```python
# Validate
print(dfAttackData.iloc[0].tolist())
```

```python
# Please note that the first column has additional space and hence It needs to be trimmed
dfAttackData['Timestamp'] = dfAttackData['Timestamp'].str.strip()
print(dfAttackData.iloc[0].tolist())
print(dfAttackData.shape)
dfAttackData.head()
```

```python
dfAttackData.to_pickle(ATTACKDATACleaned)
```

```python
# Read from raw Pickle
dfAttackDataCleaned =  pd.read_pickle(ATTACKDATACleaned)

print(dfAttackDataCleaned.shape)
dfAttackDataCleaned.head()
```
